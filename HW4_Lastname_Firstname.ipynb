**Part B**: We could instead consider describing the process of "each year is independently and identically likely to contain a heavy storm with probability $1/100$."  This would suggest a different random variable to count the number of heavy storms in a 100 year period.  Which one?

ANSWER: Binomial instead of Poisson.

```python
p = float(1/100)
dist = []

k0 = stats.binom.pmf(0,100,p)
dist.append(float(k0))
print("Probability of exactly 0 heavy rain storms:")
print(format(k0,'.4f'))
print("\n")

k1 = stats.binom.pmf(1,100,p)
dist.append(float(k1))
print("Probability of exactly 1 heavy rain storms:")
print(format(k1,'.4f'))
print("\n")

k2 = stats.binom.pmf(2,100,p)
dist.append(float(k2))
print("Probability of exactly 2 heavy rain storms:")
print(format(k2,'.4f'))
print("\n")

k3 = stats.binom.pmf(3,100,p)
dist.append(float(k3))
print("Probability of exactly 3 heavy rain storms:")
print(format(k3,'.5f'))
print("\n")

k4 = stats.binom.pmf(4,100,p)
dist.append(float(k4))
print("Probability of exactly 4 heavy rain storms:")
print(format(k4,'.6f'))
print("\n")

k5 = stats.binom.pmf(5,100,p)
dist.append(float(k5))
print("Probability of exactly 5 heavy rain storms:")
print(format(k5,'.6f'))
print("\n")

k6 = stats.binom.pmf(6,100,p)
dist.append(float(k6))
print("Probability of exactly 6 heavy rain storms:")
print(format(k6,'.7f'))
print("\n")

k7 = stats.binom.pmf(7,100,p)
dist.append(float(k7))
print("Probability of exactly 7 heavy rain storms:")
print(format(k7,'.8f'))
print("\n")

k8 = stats.binom.pmf(8,100,p)
dist.append(float(k8))
print("Probability of exactly 8 heavy rain storms:")
print(format(k8,'.9f'))
print("\n")

k9 = stats.binom.pmf(9,100,p)
dist.append(float(k9))
print("Probability of exactly 9 heavy rain storms:")
print(format(k9,'.10f'))
print("\n")

k10 = stats.binom.pmf(10,100,p)
dist.append(float(k10))
print("Probability of exactly 10 heavy rain storms:")
print(format(k10,'.11f'))

plt.bar(np.arange(len(dist)),dist, width=1, align = 'center',facecolor='blue', edgecolor='white')
plt.xticks(range(0,11))
plt.xlim([-1, 11])

plt.gca().set(title='Probability of observing k heavy rains over 100 (yearly) trials:', ylabel='Probability:', xlabel = 'Number of heavy rains:')
plt.show()
```

Suppose you are creating a model to describe how long it takes you to complete your homework assignments.  For each assignment, the time required is a random variable $X$ that returns the number of hours required for that assignment.  You decide to use the following pdf for $X$:

$$
f(x) = \begin{cases}

Find indefinite integral by applying u-substitution,

$$
u = -\frac{1}{4} \; x^2 \;,\quad \frac{du}{dx} = -\frac{1}{4} \; 2x = -\frac{1}{2} \;x\;, \quad du =  -\frac{1}{2} \;x\; dx

$$

###### **Part E.** Complete the function `integrate` below to compute an approximation to the definition integral $\int_0^{\text{ub}} f(x)dx$.  Your function should take as arguments:
1. `num_samples` -- the number of points to sample,
2. `ub` -- a number between $0$ and $1$ that is the upper bound of the integral, and
3. `isPlot` -- a Boolean flag for whether or not to display a plot of the results.

Your function should return an approximation of the definite integral by sampling $X$ from $U(0,\text{ub})$ and $Y$ from $U(0,m)$. For each random draw $(X,Y)$, define the random variable $Z$ that represents whether or not the sampled $Y$ value is less than $f(X)$ for the corresponding $X$ value.

Your function should perform three tasks: 

1. return the estimate of the value of the integral,
2. return the sample variance of the random variable $Z$ being simulated, using the np.var function, and
3. if `isPlot` is True, plot the curve $f(x)$ as well as the location of the points $(X,Y)$, colored according to whether they fall above or below the curve. 

```python
# GIVEN CODE - Do not change the inputs/outputs or your code will fail unit tests and auto-grading!
def asin(x):
    return (np.arcsin(x))/(np.pi/2-1)

def integrate(num_samples, ub, isPlot):
    #computations for estimated area here
    
    ucCount = 0 #counter to keep track of the number of random sample points that fall under the curve.
    zList = []
    xCoord = []
    yCoord = []
    xSampleCoordAbove =[]
    ySampleCoordAbove =[]
    xSampleCoordBelow =[]
    ySampleCoordBelow =[]
    
    if ub == 0:
        return ub
    
    for x in np.linspace(0,1,100):
        y = asin(x)
        xCoord.append(x)
        yCoord.append(y)
    
    for i in range(num_samples):
        
        tempX = stats.uniform.rvs(loc = 0, scale = ub, size = 1)
        tempY = stats.uniform.rvs(loc = 0, scale = 2.752, size = 1)
        fx = asin(tempX)
        
        if tempY < fx:
            
            ucCount += 1
            xSampleCoordBelow.append(tempX)
            ySampleCoordBelow.append(tempY)
            z = (tempX,tempY)
            zList.append(z)
            
        else:
            xSampleCoordAbove.append(tempX)
            ySampleCoordAbove.append(tempY)
        
    estimate = ucCount/num_samples 
    estimate = estimate*(1.376)
    sample_var = np.var(zList, ddof=1, dtype=np.float64)
    
    if isPlot:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,13)) 
        #plot here
        ax.plot(xCoord, yCoord, color ="blue")
        ax.scatter(xSampleCoordAbove, ySampleCoordAbove, 2, c = "m")
        ax.scatter(xSampleCoordBelow, ySampleCoordBelow, 2, c = "r")
        
    return estimate, sample_var 
```
